```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
prime = 61
form = lambda i, j: (i**2 + j**2) % prime
```

```python
graph = np.fromfunction(form, (prime, prime))

plt.figure(figsize=(7, 7))
plt.imshow(graph, origin='lower', cmap='viridis', interpolation='none')
```

```python
xs, ys = np.arange(0, prime), np.arange(0, prime)
xv, yv = np.meshgrid(xs, ys)
zv = form(xv, yv)

fig = plt.figure(figsize=(7, 7))
ax = fig.add_subplot(111, projection='3d')
ax.contourf(xv, yv, zv, linewidth=0, cmap='viridis')
```
